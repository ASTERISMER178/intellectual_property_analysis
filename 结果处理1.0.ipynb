```python
import pandas as pd
import numpy as np
```

```python
result1=pd.read_excel("./Data/result.xlsx")
```

```python
result1.info()
```

```python
SUMSQ=np.sum(result1["differ_rfr"]**2)
```

```python
COUNT=result1["differ_rfr"].count()
COUNT
```

```python
RSME_ALL=np.sqrt(SUMSQ/COUNT)
RSME_ALL
```

```python
result1.loc[0,"RSME_ALL"]=RSME_ALL
result1.head(5)
```

```python
TMP=result1[result1["y_test"]<=5000]
TMP.info()
```

```python
SUMSQ=np.sum(TMP["differ_rfr"]**2)
COUNT=TMP["differ_rfr"].count()
RSME_SMALL=np.sqrt(SUMSQ/COUNT)
RSME_SMALL
```

```python
result1.loc[0,"RSME_SMALL"]=RSME_SMALL
result1.head(5)
```

```python
TMP=result1[result1["y_test"]>=5000]
SUMSQ=np.sum(TMP["differ_rfr"]**2)
COUNT=TMP["differ_rfr"].count()
RSME_LARGE=np.sqrt(SUMSQ/COUNT)
result1.loc[0,"RSME_LARGE"]=RSME_LARGE
result1.head(5)
```

```python
result1.to_excel("./Data/result01.xlsx",index=False)
```
