```python
df['申请人省市代码'] = df['申请人省市代码'].map(lambda x: x.split(' ')[0])
df.replace('TW', '台湾',inplace = True)
df.replace('HK', '香港',inplace = True)
df.replace('MO', '澳门',inplace = True)
df.replace('中華民國', '台湾',inplace = True)
```

```python
df.loc[df['公开国别'] == '中国台湾', '申请人省市代码'] = '台湾'
df.loc[df['公开国别'] == '中国香港', '申请人省市代码'] = '香港'
df.loc[df['公开国别'] == '中国澳门', '申请人省市代码'] = '澳门'
selected_countries = ['中国','中国台湾', '中国香港','中国澳门']
df = df[df['公开国别'].isin(selected_countries)]
```

```python
# 使用apply和any来删除任意属性值为特定值的行
df = df[~df.apply(lambda row: any(row == "-"), axis=1)]
df = df[~df.apply(lambda row: any(row == "该数据不支持导出"), axis=1)]
```

```python
china_provinces = ["河北","山西","黑龙江","吉林","辽宁","江苏","浙江","安徽","福建","江西","山东","河南","湖北","湖南","广东","海南","四川","贵州","云南","陕西","甘肃","青海","台湾","内蒙古","广西","西藏","宁夏","新疆","北京","天津","上海","重庆","香港","澳门"]  # 包含中国各个省份的列表
df = df[df['申请人省市代码'].isin(china_provinces)]
```

```python
df.info()
df.head(5)
```

```python
for index,row in df.iterrows():
    value = row['专利价值']  # 获取当前行中 "column_name" 属性的值
    # 在这里对属性值进行操作，例如加倍
    new_value=value[1:]
    number = new_value.replace(',', '')
    # 将操作后的值赋回到当前行的属性
    df.at[index, '专利价值'] = number
df['专利价值'] = df['专利价值'].astype(np.int64)
```

```python
df['公开(公告)日'] = df['公开(公告)日'].str[:4]
df["公开(公告)日"] = pd.to_numeric(df["公开(公告)日"])
```
