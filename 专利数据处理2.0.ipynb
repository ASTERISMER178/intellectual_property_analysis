```python
df = df[~df.apply(lambda row: any(row == "该数据不支持导出"), axis=1)]
```

```python
df["权利要求数量"]=df["权利要求数量"].astype(int)
df["文献页数"]=df["文献页数"].astype(int)
df["被引用专利数量"]=df["被引用专利数量"].astype(int)
```

```python
china_provinces = ["河北","山西","黑龙江","吉林","辽宁","江苏","浙江","安徽","福建","江西","山东","河南","湖北","湖南","广东","海南","四川","贵州","云南","陕西","甘肃","青海","台湾","内蒙古","广西","西藏","宁夏","新疆","北京","天津","上海","重庆","香港","澳门"]  # 包含中国各个省份的列表
df = df[df['申请人省市代码'].isin(china_provinces)]
```

```python
df.select_dtypes(include=["object"]).describe()
```

```python
df.info()
```

```python
for index,row in df.iterrows():
    value = row['专利价值']  # 获取当前行中 "column_name" 属性的值
    # 在这里对属性值进行操作，例如加倍
    new_value=value[1:]
    number = new_value.replace(',', '')
    # 将操作后的值赋回到当前行的属性
    df.at[index, '专利价值'] = number
df['专利价值'] = df['专利价值'].astype(np.int64)
```

```python
df.info()
```

```python
df['公开(公告)日'] = df['公开(公告)日'].str[:4]
# df["公开(公告)日"] = pd.to_numeric(df["公开(公告)日"])
```

```python
df["专利价值（对数变换）"]=np.log(df["专利价值"])
```
